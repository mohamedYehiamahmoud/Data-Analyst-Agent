```python
import pandas as pd
from smolagents import tool, ToolCallingAgent, LiteLLMModel
import os


df = None
csv_file = 'Fashion_Retail_Sales.csv'
csv_file = os.path.abspath(csv_file)

task = f"""
You are an expert **Retail Business Analyst** with deep expertise in data analysis and business intelligence. Your task is to analyze the retail dataset and generate a comprehensive, professional report.

Follow these steps EXACTLY in order:

STEP 1: Load and Explore Data
- First, call load_data('{csv_file}') to load the dataset
- Use get_sample(10) to examine the first 10 rows
- Pay attention to column names, data types, and what each column represents

STEP 2: Statistical Analysis
- Call get_summary() to get descriptive statistics
- Analyze min, max, mean, and quartiles for all numeric columns
- Look for outliers, patterns, and anomalies in the data

STEP 3: Generate Comprehensive Business Report
Based on your analysis from steps 1-2, create a detailed report with these sections:

## 1. Executive Summary
- Provide 3-5 key findings that summarize the most important insights
- Focus on revenue trends, customer behavior, and product performance
- Include specific numbers and percentages where relevant

## 2. Sales Performance Analysis
- Calculate and present: total revenue, average transaction value
- Identify top-selling categories and products with their contribution percentages
- Analyze sales trends and patterns

## 3. Customer Behavior Insights
- Report number of unique customers
- Calculate average items per transaction
- Identify customer segments based on purchasing patterns
- Note any interesting customer behavior patterns

## 4. Product Performance Breakdown
- List best-performing products with their metrics
- Identify underperforming products
- Provide category-wise performance analysis
- Include revenue contribution by category

## 5. Strategic Recommendations
- Provide 3-5 specific, actionable recommendations
- Base each recommendation on data evidence from your analysis
- Explain the expected business impact of each recommendation
- Prioritize recommendations by potential impact.
"""



@tool
def load_data(file_path: str) -> str:
    """Load a CSV file for retail analysis.
    
    Args:
        file_path: Path to the CSV file to load (e.g., 'sales_data.csv')
        
    Returns:
        String confirmation message with row count, column count, and column names
    """
    global df
    try:
        df = pd.read_csv(file_path)
        return f"Loaded: {df.shape[0]} rows, {df.shape[1]} cols. Columns: {list(df.columns)}"
    except Exception as e:
        return f"Error: {str(e)}"

@tool
def get_summary() -> str:
    """Return statistical summary of the currently loaded dataset.
    
    Returns:
        String representation of pandas describe() output including count, mean, std, min, max for numeric columns
    """
    if df is None: 
        return "No data loaded"
    return df.describe(include='all').to_string()

@tool 
def get_sample(n: int = 5) -> str:
    """Return the first N rows of the currently loaded dataset.
    
    Args:
        n: Number of rows to return (default: 5)
        
    Returns:
        String representation of the first N rows of the dataframe
    """
    if df is None: 
        return "No data loaded"
    return df.head(n).to_string()

    
if os.path.exists(csv_file):
    print(f'data founded at {csv_file}')

model_id = 'ollama_chat/qwen2.5-coder:7b'

agent = ToolCallingAgent(
    tools=[load_data, get_summary, get_sample],
    model=LiteLLMModel(model_id=model_id, api_base="http://localhost:11434")
    )


report = agent.run(task)  
print(report)
```
